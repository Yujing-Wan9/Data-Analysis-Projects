```python
# =========================
# 安裝套件（Colab 第一次跑再開）
# =========================
!pip install nltk jieba transformers torch sentencepiece -q

# =========================
# 載入套件
# =========================
import nltk
import jieba
import jieba.posseg as pseg
from nltk.corpus import sinica_treebank
from transformers import pipeline

# 下載語料庫
nltk.download('sinica_treebank')

# =========================
# 先讀取 Sinica Treebank 的資料
# =========================
sinica_sents = list(sinica_treebank.sents())              # 原始斷詞結果
sinica_tagged_sents = list(sinica_treebank.tagged_sents())  # 原始詞性標註結果

```

# 1.組合原始句子並全數印出

```python
# =========================
# Part 1: 組合原始句子並全數印出
# =========================
raw_sentences = []


for sent in sinica_sents:
    raw_sentence = "".join(sent)
    raw_sentences.append(raw_sentence)

print("========== Part 1: Raw Sentences ==========")
for i, sent in enumerate(raw_sentences, 1):
    print(f"{i}. {sent}")
```

# 2. jieba 斷詞，並與 Sinica Treebank 原始斷詞結果並列印出

```python
# =========================
# Part 2: 利用 raw_sentences 做 jieba 斷詞，
# 並將 Chinese Treebank 原始斷詞結果與 jieba 斷詞結果並列印出
# =========================
print("\n========== Part 2: Segmentation Comparison ==========")

for i, raw_sent in enumerate(raw_sentences):
    jieba_result = list(jieba.cut(raw_sent))   # 用第一步組合出來的句子做斷詞
    sinica_result = sinica_sents[i]            # 對應的標準答案

    print(f"\nSentence {i+1}:")
    print("標準答案：", " / ".join(sinica_result))
    print("斷詞結果：", " / ".join(jieba_result))


```

# 3. 詞性標註，並與 Sinica Treebank 原始詞性標註結果並列印出

這個跑要3mins

```python
# =========================
# Part 3: 利用 raw_sentences 做詞性標註，
# 並將 Chinese Treebank 原始詞性標註結果與自己的結果並列印出
# =========================
print("\n========== Part 3: POS Tagging Comparison ==========")

for i, raw_sent in enumerate(raw_sentences):
    pos_answer = sinica_tagged_sents[i]  # Sinica Treebank 的標準答案
    pos_result = [(word, flag) for word, flag in pseg.cut(raw_sent)]  # 用 raw_sentences 做 POS

    print(f"\nSentence {i+1}:")
    print("標準答案：", pos_answer)
    print("詞性結果：", pos_result)


```

# 4. 實體名稱辨識，只印出有實體的句子

需跑2mins左右

```python
# =========================
# Part 4: 利用 raw_sentences 做實體名稱辨識
# 只印出有實體的句子
# =========================
print("\n========== Part 4: Named Entity Recognition ==========")

# 中文 NER 模型
ner_zh = pipeline(
    "ner",
    model="ckiplab/bert-base-chinese-ner",
    aggregation_strategy="simple"
)

for i, raw_sent in enumerate(raw_sentences):
    ner_result = ner_zh(raw_sent)   # 用第一步組合出來的句子做 NER

    identified_entities = []
    for ent in ner_result:
        entity_text = ent["word"]
        entity_label = ent["entity_group"]
        identified_entities.append((entity_text, entity_label))

    # 若該句沒有實體，則不印出
    if identified_entities:
        print(f"Sentence {i+1}: {identified_entities}")

```

# 斷詞差異統計 summary (延續第二部分

```python
# =========================
# Bonus: Segmentation Summary
# =========================
same_count = 0
diff_count = 0

for i, raw_sent in enumerate(raw_sentences):
    jieba_result = list(jieba.cut(raw_sent))
    sinica_result = sinica_sents[i]

    if jieba_result == sinica_result:
        same_count += 1
    else:
        diff_count += 1

total_count = len(raw_sentences)
accuracy = same_count / total_count * 100

print("\n========== Bonus: Segmentation Summary ==========")
print(f"Total sentences                 : {total_count}")
print(f"Exactly matched sentences       : {same_count}")
print(f"Different segmentation sentences: {diff_count}")
print(f"Exact match rate                : {accuracy:.2f}%")
```
