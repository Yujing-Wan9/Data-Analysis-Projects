Scrape popular news from Yahoo ("https://tw.news.yahoo.com/most-popular") and LTN ("https://news.ltn.com.tw/list/breakingnews/popular"). For Yahoo, implement a Python function to convert relative time strings (e.g., '小時前', '分鐘前') into absolute 'YYYY/MM/DD HH:MM' format using the current system time. For LTN, extract the full date and time for every article. Combine the results and print them as a formatted JSON string to verify data integrity. Then, connect to the local MongoDB server, clear all documents in the 'HotNews' collection of the 'NewsDB' database, and insert the newly refined records. Finally, retrieve all records from MongoDB, display them in a pandas DataFrame with Chinese column headers (標題, 網址, 日期, 來源), and provide a link to download the result as a CSV file.

```python
import requests
from bs4 import BeautifulSoup
from datetime import datetime, timedelta
import re

def convert_relative_to_absolute(rel_time_str):
    now = datetime.now()
    target_time = now

    # Extract number and unit
    match = re.search(r'(\d+)\s*(小時|分鐘)前', rel_time_str)
    if match:
        value = int(match.group(1))
        unit = match.group(2)

        if unit == '小時':
            target_time = now - timedelta(hours=value)
        elif unit == '分鐘':
            target_time = now - timedelta(minutes=value)

    return target_time.strftime('%Y/%m/%d %H:%M')

url = 'https://tw.news.yahoo.com/most-popular'
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36'}

yahoo_news = []
try:
    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.text, 'html.parser')

    # Updated selectors for Yahoo Most Popular News
    # Usually items are within a stream or a specific list
    items = soup.select('li[class*="StreamItem"]') or soup.select('div[class*="Py(14px)"]')

    if not items:
        # Fallback to a broader search for links in the main content area
        items = soup.select('h3')

    for item in items:
        title_tag = item.find('a') if item.name != 'a' else item
        if not title_tag or not title_tag.get('href'):
            continue

        title = title_tag.get_text(strip=True)
        link = title_tag['href']
        if link.startswith('/'):
            link = 'https://tw.news.yahoo.com' + link

        # Looking for time - often in a sibling or parent div
        time_text = ""
        time_elem = item.find_next('span', string=re.compile(r'前')) or item.select_one('div > span')
        if time_elem:
            time_text = time_elem.get_text(strip=True)
        else:
            time_text = "0 分鐘前"

        abs_time = convert_relative_to_absolute(time_text)

        if title and "/" in link:
            yahoo_news.append({
                'title': title,
                'url': link,
                'date': abs_time,
                'source': 'Yahoo'
            })

    # Deduplicate
    seen_links = set()
    unique_news = []
    for n in yahoo_news:
        if n['url'] not in seen_links:
            unique_news.append(n)
            seen_links.add(n['url'])
    yahoo_news = unique_news

    print(f'Successfully scraped {len(yahoo_news)} articles from Yahoo.')
    if yahoo_news:
        print(f'Example: {yahoo_news[0]}')
except Exception as e:
    print(f'An error occurred: {e}')
```

```python
import requests
from datetime import datetime
import time

# LTN's popular news is often loaded from an API endpoint
# Based on typical LTN patterns, the popular news list can be fetched via an API call.
api_url = 'https://news.ltn.com.tw/ajax/breakingnews/popular/1'
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36',
    'X-Requested-With': 'XMLHttpRequest'
}

ltn_news = []
today_str = datetime.now().strftime('%Y/%m/%d')

try:
    response = requests.get(api_url, headers=headers)
    response.raise_for_status()
    data = response.json()

    # The API usually returns a dictionary with a 'data' key containing the list of articles
    articles = data.get('data', [])
    if isinstance(articles, dict):
        # Sometimes it's a dict where keys are indices
        articles = list(articles.values())

    for article in articles:
        title = article.get('title')
        link = article.get('url')
        # LTN API often provides 'time' as MM-DD HH:MM or HH:MM
        time_val = article.get('time', '')

        if not title or not link:
            continue

        if not link.startswith('http'):
            link = 'https://news.ltn.com.tw/' + link.lstrip('/')

        # Format timestamp to YYYY/MM/DD HH:MM
        if len(time_val) == 5 and ':' in time_val:
            full_date = f"{today_str} {time_val}"
        elif '-' in time_val and ':' in time_val:
            try:
                current_year = datetime.now().year
                parsed_time = datetime.strptime(f"{current_year}-{time_val}", "%Y-%m-%d %H:%M")
                full_date = parsed_time.strftime("%Y/%m/%d %H:%M")
            except:
                full_date = f"{today_str} {time_val}"
        else:
            full_date = f"{today_str} 00:00"

        ltn_news.append({
            'title': title,
            'url': link,
            'date': full_date,
            'source': 'LTN'
        })

    print(f'Successfully scraped {len(ltn_news)} articles from LTN API.')
    if ltn_news:
        print(f'Sample record: {ltn_news[0]}')

except Exception as e:
    print(f'Error fetching LTN via API: {e}')
```

## Define Category Keywords

Create a dictionary mapping news categories to specific keywords for title-based classification.

```python
category_keywords = {
    '政治': ['總統', '外交', '政府', '選舉', '立法院', '內閣', '政黨', '兩岸', '國安'],
    '娛樂': ['明星', '演藝', '電影', '演唱會', '追星', '劇集', '綜藝', '影帝', '影后'],
    '體育': ['比賽', '冠軍', '籃球', '棒球', 'HBL', 'NBA', '賽事', '運動員', '季後賽'],
    '家庭': ['孩子', '父母', '婆婆', '媳婦', '教育', '親子', '成長', '育兒', '家事'],
    '其他': []
}

import json
print("Category Keywords Dictionary:")
print(json.dumps(category_keywords, indent=4, ensure_ascii=False))
```

## Classify and Update News Data

### Subtask:
Retrieve articles from the 'HotNews' collection, apply the keyword-based classification logic to the titles, and save the results into a new collection named 'CategorizedNews' to preserve the original records.


**Reasoning**:
I will perform the classification of news articles by retrieving them from the 'HotNews' collection, matching their titles against the `category_keywords` dictionary, and storing the categorized results into a new 'CategorizedNews' collection.



```python
from pymongo import MongoClient

# 1. Connect to MongoDB and access database
client = MongoClient('mongodb://localhost:27017/')
db = client['NewsDB']
hot_news_col = db['HotNews']

# 2. Retrieve all documents from HotNews
articles = list(hot_news_col.find({}))

# 3 & 4. Iterate and classify based on keywords
categorized_data = []
for article in articles:
    # Remove _id to avoid duplicate key errors if re-inserting, or to let MongoDB generate new ones
    # for the new collection, if the original _id is not desired.
    # For this task, we want to insert into a *new* collection, so removing _id is appropriate.
    if '_id' in article:
        del article['_id']

    title = article.get('title', '')
    assigned_category = '其他'

    for category, keywords in category_keywords.items():
        if category == '其他': # '其他' category has no keywords, it's a fallback
            continue
        if any(kw in title for kw in keywords):
            assigned_category = category
            break

    article['category'] = assigned_category
    categorized_data.append(article)

# 5 & 6. Create CategorizedNews collection and refresh data
categorized_col = db['CategorizedNews']
categorized_col.delete_many({}) # Clear existing data in the new collection

if categorized_data:
    result = categorized_col.insert_many(categorized_data)
    # 7. Print results
    print(f'Successfully classified and stored {len(result.inserted_ids)} documents in "CategorizedNews" collection.')
else:
    print('No documents found to classify.')
```

```python
import pandas as pd
from pymongo import MongoClient

# 1. Connect to MongoDB and access CategorizedNews collection
client = MongoClient('mongodb://localhost:27017/')
db = client['NewsDB']
categorized_col = db['CategorizedNews']

# 2. Retrieve all documents
categorized_docs = list(categorized_col.find({}, {'_id': 0}))
df_categorized = pd.DataFrame(categorized_docs)

# 3. Print a formatted summary of category distribution
if not df_categorized.empty:
    category_counts = df_categorized['category'].value_counts()
    print('--- News Category Distribution Summary ---')
    for category, count in category_counts.items():
        print(f'{category}: {count} articles')

    # 4. Display the first few records to confirm population
    print('\n--- Categorized News Data Preview ---')
    display(df_categorized.head())
else:
    print('No data found in CategorizedNews collection.')
```
